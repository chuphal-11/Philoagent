```python
import os
from groq import Groq
from dotenv import load_dotenv
load_dotenv()
# Initialize the client (Make sure GROQ_API_KEY is in your environment variables)
# or use: client = Groq(api_key="your_api_key_here")
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

def check_groq_access():
    print("--- 1. Checking Model Availability ---")
    try:
        models = client.models.list()
        # Common 70B IDs to look for
        target_models = ["llama3-70b-8192", "llama-3.3-70b-versatile"]
        
        available_ids = [m.id for m in models.data]
        found = [m for m in target_models if m in available_ids]
        
        if found:
            print(f" Found access to: {', '.join(found)}")
        else:
            print(" Llama 70B not found in your model list.")
            print(f"Available models: {available_ids[:5]}... (and more)")
    except Exception as e:
        print(f" Connection Error: {e}")
        return

    print("\n--- 2. Testing Chat Completion ---")
    # Using the most common current 70B model ID
    model_to_test = "llama-3.3-70b-versatile" if "llama-3.3-70b-versatile" in available_ids else "llama3-70b-8192"
    print(f"Testing chat completion with model: {model_to_test}")
    try:
        completion = client.chat.completions.create(
            model=model_to_test,
            messages=[{"role": "user", "content": "Hello, are you active?"}],
        )
        print(f" Success! Response from {model_to_test}:")
        print(f"Assistant: {completion.choices[0].message.content}")
    except Exception as e:
        print(f" API Request Failed: {e}")

if __name__ == "__main__":
    check_groq_access()
```
